```python
import requests

url = "https://cine5-a2ia-12.onrender.com/recommend"

payload = {
    "movie_ids": [8, 2, 3, 4, 5]

}

try:
    res = requests.post(url, json=payload, timeout=20)
    print("Status:", res.status_code)

    try:
        print("resulat", res.json())
    except ValueError:
        print("resulat:", res.text)

except Exception as e:
    print("Error:", e)
```
